Escape Filter for Compton Electrons in CCD
=====

1. Introduction
----
For the past decade, Compton imaging systems have been developed based on position sensitive semiconductor detectors for use in Homeland Security and Nuclear Safeguards applications. The conventional Compton imaging is using the coincidence events in the detectors and back-project a cone surface which indicating the possible gamma-ray direction, as shown below.

```python
from IPython.core.display import *
Image(filename="images/Compton.png")
```

The noval Compton imager developed at LBNL uses high-resolution scientific charge-coupled device to measure the trajectory of Compton electron. With the information of the electron, the back-projected cone can be reduced to an arc. So there will be a gain in sensitivity.

```python
Image(filename="images/ETCI.jpg")
```

But there's a chance that the electron escapes the detector so that we are not able to measure the total energy of the electrons. Consequently, this false energy will add noise into our image. So the problem I am trying to solve here is to find a filter for electrons escaping the detector volumn.

There are about 250,000 tracks generated by GEANT4 simulation, in which we know whether the electron escapes or not. So I am gonna use a data driven method to design the filter.

2. Background
---

### Data conversion
The Geant4 simulation data are save as MATLAB .mat files. And there are a lot more information than the track saved there. So the first thing is to transfer the .mat data into hdf5. It is done in MATLAB. To read the data, PyTable is used. After loading into python, the events are saved as .pkl file. 

### Feature extraction
The features that are important to an electron track are:

- Number of pixels the track occupies
- The total energy deposition
- The maximum energy depostion in one pixel
- The Bragg peak energy with a 5 \times 5 filter
- Maximum value after Laplacian operator

Parallelism is implemented here since same operations are done to the tracks.

### Classifiers
Since it is a binary classification, the classifiers will be used are:

- Logistic regression
- LDA
- QDA

Cross validation with fold=5 are used to validate the model.

### Parameter selection

In Logistic regression, cross validation is done manually. And the coefficients are saved for each time. For each coefficient, the higher mean/std, the more important this coefficient is.

```python
# don't need to run this if already pickled.
def to_dict(fname):
	import tables as tb
	import numpy as np
	import pandas as pd
	import cPickle as pickle
	from collections import OrderedDict 

	data = tb.openFile(fname)
	cmd_0 = 'data.root.'
	event = OrderedDict()
	for i in xrange(200000):
		cmd_track = cmd_0 + 'track.d' + str(i) + '.read()'
		cmd_escape = cmd_0 + 'escape.d' + str(i) + '.read()'
		track = eval(cmd_track)[0]
		escape = eval(cmd_escape)[0]
		event['d'+str(i)] = [track, escape[0]]
	event
	pickle.dump(event, open('event_200k.pkl', 'wb'))
	return event
fname = 'electronTrack_200k.h5'
event = to_dict(fname)
```

```python
%%px 
def num_pixel(track):
    thres = 0.0055
    return numpy.sum(track>thres)

def Bragg_E(track):
    PSF = numpy.ones((5,5))
    PSF[0,0] = 0
    PSF[0,-1] = 0
    PSF[-1,0] = 0
    PSF[-1,-1] = 0
    c = convolve2d(track, PSF, mode='same')
    return c.max()

def max_E(track):
    return track.max()

def total_E(track):
    return track.sum()

def highpass_max(track):
    PSF = numpy.zeros((3,3))
    PSF[0,1] = -1
    PSF[1,0] = -1
    PSF[1,-1] = -1
    PSF[1,1] = 4
    PSF[-1,1] = -1
    hp = convolve2d(track, PSF, mode='same')
    return hp.max()
    
def feature_extract(key_names):
    '''list of key names
    '''
    result=[]
    for key in key_names:
        track = event_engin[key][0]
        escape = event_engin[key][1]
        
        num_p = num_pixel(track)
        Bragg = Bragg_E(track)
        Emax = max_E(track)
        Etot = total_E(track)
        E_hp = highpass_max(track)
        result.append((num_p, Bragg, Emax, Etot, E_hp, escape))
    return result
feature_table = feature_extract(key_list)
```

```python
def print_cv_summary(model, xx, yy, cv, printFlag=True):
    scores = cross_val_score(model, xx, yy, cv=cv, n_jobs=-1, scoring='roc_auc')
    if printFlag:
        print("ROC curve: mean: {:3f}, stdev: {:3f}".format(\
              np.mean(scores), np.std(scores)))
    else:
        pass
    ROC = np.mean(scores)
    ROC_std = np.std(scores)
    scores = cross_val_score(model, xx, yy, cv=cv, n_jobs=-1, scoring='precision')
    if printFlag:
        print("Precision(TP/(TP+FP)): mean: {:3f}, stdev: {:3f}".format(\
              np.mean(scores), np.std(scores)))  
    else:
        pass
    prec = np.mean(scores)
    prec_std = np.std(scores)
    
    scores = cross_val_score(model, xx, yy, cv=cv, n_jobs=-1, scoring='accuracy')
    if printFlag:
        print("Accuracy((TP+TN)/All): mean: {:3f}, stdev: {:3f}".format(\
          np.mean(scores), np.std(scores)))  
    else:
        pass
    accu = np.mean(scores)
    accu_std = np.std(scores)
#    res = [('col', list(xx.columns.values)), ('ROC', ROC), ('ROC_std', ROC_std), ('prec', prec ), ('prec_std', prec_std)]
    res_values = [list(xx.columns.values), ROC, ROC_std, prec, prec_std, accu, accu_std]
    return  res_values
```

## parameter selection

### look at ROC AUC for different combination of features

```python
parameters_idx = range(5)
selection_list = []
for npara in xrange(5,2,-1):
    for para in combinations(parameters_idx, npara):
        para_list = [X_cut.columns[x] for x in para]
        selection_list.append(print_cv_summary(clf_2, X_cut.ix[:,para_list], Y_cut, cv=cross_validation.KFold(len(Y_cut),5)))
```

### mannual cross validation. look at the std vs true_value

```python
nfold = 5
X_folds = np.array_split(X, nfold)
Y_folds = np.array_split(Y, nfold)
result = []
for k in range(nfold):
    X_train = list(X_folds)
    X_test = X_train.pop(k)
    X_train = np.concatenate(X_train)
    Y_train = list(Y_folds)
    Y_test = Y_train.pop(k)
    Y_train = np.concatenate(Y_train)
    model = clf_2.fit(X_train, Y_train)
    betas = list(model.coef_[0])
    intercept = list(model.intercept_)
    betas.extend(intercept)
    result.append(betas)
```

```python
df = pd.DataFrame(result)
coeff_std = df.std()
coeff_mean = abs(df.mean())
coeff_mean/coeff_std
```
